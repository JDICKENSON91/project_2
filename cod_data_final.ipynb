## Cause of death due to injury

```python
cod_injury_stats = "Resources/cod_injury_stats.csv"
```

```python
cod_stats_df = pd.read_csv(cod_injury_stats, low_memory=False, error_bad_lines=False)
```

```python
cod_stats_df
```

```python
clean_cod= cod_stats_df.dropna()
clean_cod
```

```python
Avg=clean_cod.mean(axis=1)
Avg
```

```python
avg_df= pd.DataFrame(Avg)
avg_df
```

```python
clean_cod= pd.merge(clean_cod, avg_df, left_index=True, right_index=True)
clean_cod
```

```python
clean_cod_2 = clean_cod.rename(columns={"Country Name": "country_name", "Country Code": "country_code", 0:"average_per"})
```

```python
columns= ["country_code", "average_per"]

injury= clean_cod_2[columns].copy()
injury
```

```python
injury.index.rename("id", inplace=True)
injury
```
